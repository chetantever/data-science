Matplotlib - plotting library in python

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional Method
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('X Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1) #dimension
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
## Object Oriented method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```
