### Decision Trees

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('kyphosis.csv')
df.head()
```

```python
df.info()
```

```python
sns.pairplot(df,hue='Kyphosis')
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Kyphosis',axis=1)
y = df['Kyphosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

### Random Forests 

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train,y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,rfc_pred))
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
loans = pd.read_csv('loan_data.csv')
loans.head()
```

```python
loans.info()
```

```python
loans.describe()
```

```python
#Exploratory Data Analysis
plt.figure(figsize=(10,6))
loans[loans['credit.policy']==1]['fico'].hist(bins=35,color='blue',label='Credict Policy = 1',alpha=0.6)
loans[loans['credit.policy']==0]['fico'].hist(bins=35,color='red',label='Credict Policy = 0',alpha=0.6)
plt.legend()
plt.xlabel('FICO score')

#less than ~660 FICO score, credict policy risk is high
```

```python
#Exploratory Data Analysis
plt.figure(figsize=(10,6))
loans[loans['not.fully.paid']==1]['fico'].hist(bins=35,color='blue',label='Not fully paid = 1',alpha=0.6)
loans[loans['not.fully.paid']==0]['fico'].hist(bins=35,color='red',label='Not fully paid = 0',alpha=0.6)
plt.legend()
plt.xlabel('FICO score')
```

```python
plt.figure(figsize=(11,7))
sns.countplot(x='purpose',hue='not.fully.paid',data=loans,palette='Set1')
```

```python
sns.jointplot(x='fico',y='int.rate',data=loans,color='purple')
```

```python
plt.figure(figsize=(11,7))
sns.lmplot(y='int.rate',x='fico',data=loans,hue='credit.policy',col='not.fully.paid',palette='Set1')
```

```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
final_data.info()
```

```python
final_data.head()
```

#### Decision Trees

```python
from sklearn.model_selection import train_test_split
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

#### Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train,y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,rfc_pred))
```

```python
print(classification_report(y_test,rfc_pred))
```
